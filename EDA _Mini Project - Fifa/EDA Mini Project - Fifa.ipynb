```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option("display.max_columns",None)
```

```python
# Loading Data Set
df1=pd.read_csv("fifa.csv")
```

```python
# First few rows of data set
df1.head(5)
```

```python
# Number of rows and columns present in data set
df1.shape
```

```python
# Information of data set (column name,non-null-value counts,Datatype of particular column)
df1.info()
```

```python
# Total number of null values present in data set
df1.isnull().sum().sum()
```

```python
# Null values present in column in percent
df1.isnull().sum()/len(df1)*100
```

```python
# There are around 93% entries are null in a column named 'Loaned From'. So we can drop that column.
del df1["Loaned From"]
```

```python
# Shape of data after deleting Loaned from column.
df1.shape
```

```python
# removing "€" ,replacing "M" with 000000 and "K" with 000 also changing of dtype to float.
df1["Value"] = df1["Value"].str.replace("€","").str.replace("M","000000").str.replace("K","000").str.replace(".","").astype("float")
```

```python
# removing "€" ,replacing "M" with 000000 and "K" with 000 also changing of dtype to float.
df1["Wage"] = df1["Wage"].str.replace("€","").str.replace("M","000000").str.replace("K","000").str.replace(".","").astype("float")
```

```python
# Extracted year value and changing dtype to float
df1["Joined"] = df1["Joined"].str[-4:].astype("float")
```

```python
# Converting dtype to datetime
df1["Contract Valid Until"]=pd.to_datetime(df1["Contract Valid Until"])
```

```python
# Replacing (') with (.) and converting dtype to float
df1["Height"]=df1["Height"].str.replace("'",".").astype("float")
```

```python
# Replacing (lbs) with null and converting dtype to float
df1["Weight"]=df1["Weight"].str.replace("lbs","").astype("float")
```

```python
# Converting to Float after getting rid of currency symbol and suffix.
df1["Release Clause"]=df1["Release Clause"].str.replace("€","").str.replace("M","000000").str.replace("K","000").str.replace(".","").astype("float")
```

```python
df1.isnull().sum()
```

```python
df1 = df1.dropna(thresh=20)
```

```python
df1["Position"].dropna(inplace=True)
```

```python
df1.isnull().sum()
```

```python
df1 = df1.drop(index=df1[df1["Position"].isnull()].index)
```

# Exporatory Data Analysis

```python
#1.Plot the distribution of Overall rating for all players. 
sns.kdeplot(df1["Overall"])
plt.title("Distribution of Overall rating for all players")
plt.show()
```

```python
#2.Generate pair plots for the following variables:(Overall, Value, Wage, International Reputation, Height, Weight, Release Clause)
sns.pairplot(df1[["Overall", "Value", "Wage", "International Reputation", "Height", "Weight", "Release Clause"]])
plt.show()
```

```python
#3.Generate a table containing the top 20 players ranked by Overall score and whose contract expires in 2020.
#a)What would the average wage for this set of players be?
#b)What is the average age?
#c)Is there a correlation between the Overall rating and Value for these players?
```

```python
df3=df1[df1["Contract Valid Until"].dt.year==2020].sort_values(by="Overall",ascending=False).head(20)
```

```python
print("The average wage for this set of players is",df3["Wage"].mean())
```

```python
print("The average age for this set of players is",df3["Age"].mean())
```

```python
print("The correlation between the variables Overall rating and Value for these players is",df3["Overall"].corr(df3["Value"]))
```
